```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 6, 76, 32, 21])
plt.plot(x, y, 'rd:')
plt.ylabel('Random Values')
plt.xlabel('Pointers')
plt.title('Plotting a random graph')
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 6, 76, 32, 21])
plt.plot(x, y, 'gd:')
plt.ylabel('Random Values')
plt.xlabel('Pointers')
plt.title('Plotting a random graph')

plt.axis([0, 9, 0, 200])
# 0,9 for x axis and 0,200 for y axis
plt.show()
```
