```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
N = 50
x = np.array([1, 2, 4, 5])
y = np.array([23, 12, 45, 32])

plt.plot(x, y, color = 'orange', marker = 'o')
plt.show()
```

```python
plt.plot(x, y, 'or')
```

```python
plt.plot(x, y, 'b+')
```

```python
plt.plot(x, y, 'b--')
```

```python
'''
r for red,
o for marker,
- for line pattern.
'''

plt.plot(x, y, 'ro-')
```

```python
plt.plot(x, y, 'g--d')
```

```python
plt.plot(x, y, 'k^')
```

```python
plt.plot(x, y, 'k^:')
```

```python
plt.plot(x, y, 'r^-')
```

```python
plt.plot(x, y, 'gd:', linewidth = 8)
```
